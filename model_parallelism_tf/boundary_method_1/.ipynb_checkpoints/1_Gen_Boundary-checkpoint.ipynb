# Util Functions

```python
def find_boundaries_1(img):
    img1 = np.zeros((img.shape[0] + 2, img.shape[1] + 2))
    img1[1:-1, 1:-1] = img
    boundary = find_boundaries(img1)
    return boundary[1:-1, 1:-1]

def calc_angle(x, y, x0, y0):
    if y == y0:
        alpha = 90 if x >= x0 else 270
        return alpha
    tan_alpha = np.abs((1.0*(x - x0)) / (1.0*(y - y0)))
    alpha = np.arctan(tan_alpha)*180/np.pi
    if x > x0:
        if y < y0:
            alpha = 180 - alpha
    else:
        if y > y0:
            alpha = 360 - alpha
        else:
            alpha = 180 + alpha
    if alpha == 360:
        alpha = 0
    return alpha

def to_list(x):
    if hasattr(x, '__len__'):
        return [to_list(l) for l in x]
    else:
        return x

def create_two_dimension_list(n):
    res = []
    for _ in range(n):
        res.append([])
    return res

def find_largest_connected_element(lake_boundary):
    visited, label = measurements.label(lake_boundary)
    area = measurements.sum(lake_boundary, visited,
                            index = np.arange(label + 1))
    largest_element = np.argmax(area)
    return np.where(visited==largest_element, 1, 0)

def create_empty_list(n, default_value=None):
    res = []
    for _ in range(n):
        if default_value is not None:
            res.append(default_value.copy())
    return res

def distance(p1, p2):
    return (p1[0] - p2[0])**2 + (p1[1] - p2[1])**2

def sortby(a, b):
    return (a[0] > b[0]) or (a[0] == b[0] and a[1] > b[1])
key = functools.cmp_to_key(sortby)

def find_appropriate_coor(list_angles_candidates_id, diff_arr, pos, return_id=False):
    record = []
    for id, angle in zip(list_angles_candidates_id, diff_arr):
        record.append([angle, [pos[0][id], pos[1][id]], id])
    
    distances = []
    origin = record[0][1]
    for elem in record:
        p = elem[1]
        distances.append([distance(p, origin), elem])
    distances = sorted(distances, key=itemgetter(0))
    
    type_id = 0
    distances[0][1].append(type_id)
    for i in range(1, len(distances)):
        if distances[i][0] - distances[i - 1][0] > 100:
            type_id += 1
        distances[i][1].append(type_id)
    
    group_by_types = create_empty_list(type_id + 1, [])
    for elem in distances:
        record_i = elem[1]
        t_id = record_i[-1]
        group_by_types[t_id].append(record_i[:-1])
    
    res = []
    res_id = []
    for elem in group_by_types:
        elem = sorted(elem, key=itemgetter(0))
        res.append(elem[0][1])
        res_id.append(elem[0][-1])
    if return_id:
        return res, res_id
    return res

def group_by_angle(x, angle):
    res = create_empty_list(360, [])
    for p in x:
        p_angle = angle[p[0], p[1]]
        res[int(p_angle)].append(p)
    return res

def get_len_group(boundary_vectors, angle):
    group_angle = group_by_angle(boundary_vectors, angle)
    return [len(l) for l in group_angle]

def create_increase_abs_list(n):
    res = [0]
    for i in range(1, n):
        res += [-i, i]
    return res

def reduce(group_x, permanent_group, i):
    if len(permanent_group[i]) == 0:
        return []
    group = group_x[i].copy()
    return _reduce(group, permanent_group[i])

def _reduce(group, permanent_group_i):
    res = []
    for p in permanent_group_i:
        d = np.asarray([distance(p, q) for q in group])
        i = np.argmin(d)
        res += [group[i]]
        group.remove(group[i])
    return res

def enhance(group_x, permanent_group, i):
    new_group = []
    for jj in create_increase_abs_list(10):
        j = i + jj
        if j < 0:
            j += 360
        if j >= 360:
            j -= 360
        new_group += group_x[j]
    return _reduce(new_group, permanent_group[i])

def normalize(x, permanent_group, angle):
    group_x = group_by_angle(x, angle)
    res = []
    for i in range(len(group_x)):
        if len(group_x[i]) == len(permanent_group[i]):
            res += group_x[i]
        else:
            if len(group_x[i]) < len(permanent_group[i]):
                normalized_list = enhance(group_x, permanent_group, i)
                res += normalized_list
            else:
                normalized_list = reduce(group_x, permanent_group, i)
                res += normalized_list
    return res

def find_most_frequence_range(data, threshold):
    values, counts = np.unique(data, return_counts=True)
    most_frequence_values = []
    for value, count in zip(values, counts):
        if count >= threshold:
            most_frequence_values.append(value)
    return np.asarray(most_frequence_values)

def remove_outliers(data, threshold=15):
    data1 = data.copy()
    most_frequence_range = find_most_frequence_range(data1, threshold)
    default_value = np.mean(most_frequence_range)
    data1[(data1 < most_frequence_range[0]) | (data1 > most_frequence_range[-1])] = default_value
    return data1

def remove_outliers_test(data_train, data_test, threshold=15):
    data1 = data_test.copy()
    most_frequence_range = find_most_frequence_range(data_train, threshold)
    default_value = np.mean(most_frequence_range)
    data1[(data1 < most_frequence_range[0]) | (data1 > most_frequence_range[-1])] = default_value
    return data1
```

# BoundaryVectorizer Class

```python
class BoundaryVectorizer:
    def __init__(self, n_lines, list_filenames, permanent_water_pixels):
        self.list_filenames = list_filenames
        img = restore_data(self.list_filenames[0])
        self.height, self.width = img.shape[:2]
        self.angle = np.zeros((self.height, self.width))
        self.center_x = self.height // 2
        self.center_y = self.width // 2
        self.permanent_water_pixels = permanent_water_pixels
        
        for i in range(self.height):
            for j in range(self.width):
                self.angle[i, j] = calc_angle(i, j, 256, 256)
        
        self.n_lines = n_lines
        self.len = len(list_filenames)
        self.line_angles = np.linspace(0, 360, self.n_lines + 1)[:-1]
        self.list_int_line_angles = create_two_dimension_list(360)
        for i, angle in enumerate(self.line_angles):
            self.list_int_line_angles[int(angle)].append(i)
        
        self.n_points_per_line = []
        self.line_idx_of_points = []
        self.permanent_boundary_vector = self.get_permanent_boundary_vector()
        self.permanent_group = group_by_angle(self.permanent_boundary_vector, self.angle)
        self.n_points = len(self.permanent_boundary_vector)
        print('n_points =', self.n_points)
        
        self.boundary_vectors = [None]*self.len
    
    
    def __len__(self):
        return self.len
    
    def get_permanent_boundary_vector(self):
        return self._vectorize_boundary_1(self.permanent_water_pixels, permanent=True)
    
    
    def __getitem__(self, idx):
        if self.boundary_vectors[idx] is not None:
            return self.boundary_vectors[idx]
        
        boundary_vector = self._vectorize_boundary(idx)
        boundary_vector = self._normalize(boundary_vector)
        self.boundary_vectors[idx] = boundary_vector
        if len(boundary_vector) != self.n_points:
            print(idx)
        return boundary_vector
    
    def _normalize(self, boundary_vector):
        return normalize(boundary_vector, self.permanent_group, self.angle)
    
    def _vectorize_boundary(self, idx):
        img = restore_data(self.list_filenames[idx])
        img = np.bitwise_or(img, permanent_water_pixels)
        return self._vectorize_boundary_1(img)
    
    def _vectorize_boundary_1(self, img, permanent=False):
        boundary = find_boundaries_1(img)
        boundary = find_largest_connected_element(boundary)
        pos = np.where(boundary)
        list_point_angles = np.zeros(len(pos[0]))
        for i, (x, y) in enumerate(zip(*pos)):
            list_point_angles[i] = self.angle[x, y]
                
        list_int_angles_id = create_two_dimension_list(360)
        list_int_angles_value = create_two_dimension_list(360)
        for i, angle in enumerate(list_point_angles):
            list_int_angles_id[int(angle)].append(i)
            list_int_angles_value[int(angle)].append(angle)
        
        boundary_vector = []
        for i, line_angle in enumerate(self.line_angles):
            int_line_angle = int(line_angle)

            for j in range(-1, 2):
                next_int_line_angle = int_line_angle + j
                if next_int_line_angle < 0:
                    next_int_line_angle += 360
                elif next_int_line_angle >= 360:
                    next_int_line_angle -= 360
                list_angles_candidates_id = list_int_angles_id[next_int_line_angle]
                list_angles_candidates_value = list_int_angles_value[next_int_line_angle]
                if len(list_angles_candidates_id) == 0:
                    if permanent:
                        self.n_points_per_line.append(0)
                    continue
                list_angles_candidates_value = np.asarray(list_angles_candidates_value)

                if len(list_angles_candidates_id) > 1:
                    diff_arr = np.abs(list_angles_candidates_value - line_angle)
                    list_new_points = find_appropriate_coor(list_angles_candidates_id, diff_arr, pos, permanent)
                    if permanent:
                        boundary_vector += list_new_points[0]
                        self.line_idx_of_points += list_new_points[1]
                    else:
                        boundary_vector += list_new_points
                    if permanent:
                        self.n_points_per_line.append(len(list_new_points))
                else:
                    point_idx = list_angles_candidates_id[0]
                    boundary_vector += [[pos[0][point_idx], pos[1][point_idx]]]
                    if permanent:
                        self.n_points_per_line.append(1)
        boundary_vector = list(set(tuple(l) for l in boundary_vector))
        boundary_vector = sorted(boundary_vector, key=itemgetter(0,1))
        return to_list(boundary_vector)
        
    def get_boundary_vectors(self):
        for i in range(len(self.boundary_vectors)):
            if self.boundary_vectors[i] is None:
                self.boundary_vectors[i] = self.__getitem__(i)
        return np.asarray(self.boundary_vectors)
```

```python
def find_border(data_points):
    x = data_points[:, 0]
    y = data_points[:, 1]
    return x.min(), y.min(), x.max(), y.max()

def calc_type(data_points, diagonal_line):
    point_types = []
    a, b, c = diagonal_line
    for data_point in data_points:
        x, y = data_point
        point_type = 1 if a*x + b*y + c > 0 else 0
        point_types.append(point_type)
    return point_types

def get_line(p):
    x1, y1, x2, y2 = p
    a = y1 - y2
    b = x2 - x1
    c = x1*y2 - x2*y1
    return a, b, c
```

```python
def sort_points(data_points, order_sample):
    diagonal_points = find_border(data_points)
    diagonal_line = get_line(diagonal_points)
    point_types = calc_type(data_points, diagonal_line)
    data_points = data_points.tolist()
    for data_point, order, point_type in zip(data_points, order_sample, point_types):
        data_point.append(order)
        data_point.append(point_type)
    data_points = sorted(data_points, key=itemgetter(3,2))
    return data_points
```
